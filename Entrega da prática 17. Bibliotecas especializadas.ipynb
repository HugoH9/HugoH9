```python
# Importar numpy e utilizar um alias para poder utilizar os seus recursos
import numpy as np

# Criar um array de elementos (do 10 a 100)
array1 = np.arange(10, 101)

# Inverter o array
array1_invertido = array1[::-1]

# Criar um array 3x3 com valores de 0 a 8
array2 = np.arange(9).reshape(3, 3)

# Criar um array identidade de 6x6
array_identidade = np.eye(6)

# Criar um array com valores ao acaso com forma 3x3x3
array_aleatorio = np.random.random((3, 3, 3))

# Criar dois arrays, com valores ao acaso, A e B
A = np.random.random((3, 3))
B = np.random.random((3, 3))

# Verificar se os arrays A e B são iguais
sao_iguais = np.array_equal(A, B)

# Alternativamente, você pode verificar a igualdade elemento a elemento usando np.all
sao_iguais_elemento_a_elemento = np.all(A == B)

# Outra opção é verificar se os elementos estão aproximadamente próximos dentro de uma tolerância
sao_aproximadamente_iguais = np.allclose(A, B)

# Imprimir os resultados
print("Array de 10 a 100:", array1)
print("Array invertido:", array1_invertido)
print("Array 3x3 com valores de 0 a 8:")
print(array2)
print("Array identidade de 6x6:")
print(array_identidade)
print("Array aleatório 3x3x3:")
print(array_aleatorio)
print("Os arrays A e B são iguais:", sao_iguais)
print("Os arrays A e B são iguais elemento a elemento:", sao_iguais_elemento_a_elemento)
print("Os arrays A e B são aproximadamente iguais:", sao_aproximadamente_iguais)

```

```python
import pandas as pd

# Load 05_05_imdb_títulos.csv as pandas dataframe and show first 5 records
df_titulos = pd.read_csv("05_05_imdb_títulos.csv")
print(df_titulos.head(5))

# Load 05_05_imdb_elencos.csv as pandas dataframe and show first 5 records
df_elencos = pd.read_csv("05_05_imdb_elencos.csv")
print(df_elencos.head(5))

# Show number of records in titles dataframe
print("Number of records in titles dataframe:", len(df_titulos))

# Show number of records in cast dataframe
print("Number of records in cast dataframe:", len(df_elencos))

# Show 5 oldest titles
print("5 oldest titles:")
print(df_titulos.sort_values("year").head(5)["title"])

# Show films containing the word "Drácula" and count number of films
dracula_films = df_titulos[df_titulos["title"].str.contains("Drácula")]
print("Number of films containing the word 'Drácula':", len(dracula_films))

# Show 10 most common titles
print("10 most common titles:")
print(df_titulos["title"].value_counts().head(10))

# Show first film titled "Romeo and Juliet"
romeo_juliet = df_titulos[df_titulos["title"] == "Romeo and Juliet"]
print("First film titled 'Romeo and Juliet':")
print(romeo_juliet.sort_values("year").head(1)["year"])

# List all films containing the word "Exorcist", ordered from oldest to newest
exorcist_films = df_titulos[df_titulos["title"].str.contains("Exorcist")]
print("Films containing the word 'Exorcist':")
print(exorcist_films.sort_values("year"))

# Show number of films made in 1950
print("Number of films made in 1950:", len(df_titulos[df_titulos["year"] == 1950]))

# Show number of films made from 1950 to 1959 (inclusive)
print("Number of films made from 1950 to 1959:", len(df_titulos[(df_titulos["year"] >= 1950) & (df_titulos["year"] <= 1959)]))

# Show all roles in "The Godfather" trilogy, grouping by role and showing the number of times the role appeared
godfather_titles = df_titulos[df_titulos["title"].str.contains("The Godfather")]
godfather_cast = df_elencos[df_elencos["tconst"].isin(godfather_titles["tconst"])]
print("Roles in 'The Godfather' trilogy:")
print(godfather_cast["nconst"].value_counts())

# Show complete cast of "Drácula" (1958), sorted by "n" value
dracula_1958 = df_titulos[(df_titulos["title"] == "Drácula") & (df_titulos["year"] == 1958)]
dracula_cast = df_elencos[df_elencos["tconst"].isin(dracula_1958["tconst"])].sort_values("n")
print("Complete cast of 'Drácula' (1958):")
print(dracula_cast)

# Show number of roles played by "Bruce Wayne" (including "Young Bruce Wayne")
bruce_wayne = df_elencos[df_elencos["primaryName"].str.contains("Bruce Wayne")]
print("Number of roles played by 'Bruce Wayne':", len(bruce_wayne))

# Show number of roles played by "Robert de Niro"
robert_de_niro = df_elencos[df_elencos["primaryName"] == "Robert De Niro"]
print("Number of roles played by 'Robert de Niro':", len(robert_de_niro))

charlton_heston = df_elencos[(df_elencos["name"] == "Charlton Heston") & (df_elencos["n"] == 1) & (df_elencos["year"] // 10 == 196)]
print("Lead roles (n=1) for Charlton Heston in the 60's, sorted by year (descending):")
print(charlton_heston.sort_values("year", ascending=False))

# Show number of roles for actors in the 50's
roles_actors_50s = df_elencos[(df_elencos["type"] == "actor") & (df_elencos["year"] // 10 == 195)]
print("Number of roles for actors in the 50's:", len(roles_actors_50s))

# Show number of roles for actresses in the 50's
roles_actresses_50s = df_elencos[(df_elencos["type"] == "actress") & (df_elencos["year"] // 10 == 195)]
print("Number of roles for actresses in the 50's:", len(roles_actresses_50s))
```
