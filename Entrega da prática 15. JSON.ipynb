```python
# Criando as listas
experiencia_chefe = ["Gestão", "Finanças", "Bases de dados"]
experiencia_helena = ["JavaScript", "Python"]
experiencia_luis = ["Django", "Flask", "Pyramid"]

# Criando os dicionários para cada funcionário
chefe_projeto = {
    "Nome": "João",
    "Idade": 28,
    "Experiência": experiencia_chefe,
    "Residência": "Porto",
    "HorasProjeto": 3500
}

helena = {
    "Nome": "Helena",
    "Idade": 26,
    "Experiência": experiencia_helena,
    "Residência": "Faro",
    "HorasProjeto": 500
}

luis = {
    "Nome": "Luis",
    "Idade": 31,
    "Experiência": experiencia_luis,
    "Residência": "Lisboa",
    "HorasProjeto": 1100
}

# Criando o dicionário principal
data = {
    "chefe_projeto": chefe_projeto,
    "funcionários": [helena, luis]
}

# Comprovando o tipo e mostrando na tela
print(type(data))

# Calculando as horas
horas_chefe = data['chefe_projeto']['HorasProjeto']
horas_funcionarios = sum(funcionario['HorasProjeto'] for funcionario in data['funcionários'])
horas_total = horas_chefe + horas_funcionarios

print("Horas total:", horas_total)
```

```python
import json

# Criando as listas
experiencia_chefe = ["Gestão", "Finanças", "Bases de dados"]
experiencia_helena = ["JavaScript", "Python"]
experiencia_luis = ["Django", "Flask", "Pyramid"]

# Criando os dicionários para cada funcionário
chefe_projeto = {
    "Nome": "João",
    "Idade": 28,
    "Experiência": experiencia_chefe,
    "Residência": "Porto",
    "HorasProjeto": 3500
}

helena = {
    "Nome": "Helena",
    "Idade": 26,
    "Experiência": experiencia_helena,
    "Residência": "Faro",
    "HorasProjeto": 500
}

luis = {
    "Nome": "Luis",
    "Idade": 31,
    "Experiência": experiencia_luis,
    "Residência": "Lisboa",
    "HorasProjeto": 1100
}

# Criando o dicionário principal
data = {
    "chefe_projeto": chefe_projeto,
    "funcionários": [helena, luis]
}

# Convertendo dicionário para JSON String
json_string = json.dumps(data, indent=4)

# Comprovando o tipo da variável json_string
print(type(json_string))

# Imprimindo dados
print(json_string)

# Resposta à pergunta: Para que serve este formato?
print("O formato JSON (JavaScript Object Notation) é amplamente utilizado para armazenar e trocar dados estruturados entre sistemas. Ele é um formato de texto leve e fácil de ler que permite representar informações de forma organizada. Serve para serializar dados e transmiti-los pela rede, armazenar configurações, trocar informações entre diferentes linguagens de programação e muito mais. É uma forma padronizada de representar dados, o que o torna universalmente compreensível e fácil de usar em diversas aplicações.")
```

```python
import json

# Descodificando o JSON
data = """
{
    "Superman": {
        "Poderes": ["Super força", "Super velocidade", "Visão de raio X"],
        "Idade": 40
    },
    "Batman": {
        "Poderes": ["Intelecto sobre-humano", "Raios nos olhos"],
        "Idade": 350
    },
    "Wonder Woman": {
        "Poderes": ["Força sobre-humana", "Velocidade sobre-humana"],
        "Idade": 35
    }
}
"""

data = json.loads(data)

# Corrigindo os erros
data["Superman"]["Poderes"].append("Voar")
data["Batman"]["Idade"] = 35
data["Batman"]["Poderes"].remove("Raios nos olhos")
data["Wonder Woman"]["Poderes"].append("Laço da verdade")

# Recodificando para JSON String
data = json.dumps(data)

# Mostrando o tipo
print(f"Tipo: {type(data)}")

# Mostrando os dados
print(data)
```

```python
import json

super_herois = {
    "Superman": {"poderes": ["Super força", "Visão de raio X", "Voar"]},
    "Batman": {"poderes": ["Inteligência", "Riqueza", "Artimanhas"]},
    "Wonder Woman": {"poderes": ["Força", "Agilidade", "Laço da verdade"]}
}

membros_ativos = ["Superman", "Wonder Woman"]
membros_inativos = ["Batman"]

super_herois["membrosAtivos"] = membros_ativos
super_herois["membrosInativos"] = membros_inativos

json_string = json.dumps(super_herois, indent=4)
print(json_string)
print(type(json_string))
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)

data = response.json()

print("Tipo de dados recebido:", type(data))
print("Dados recebidos:", data)
print("Número de pessoas na estrutura:", len(data))

for pessoa in data:
    print("Nome:", pessoa['name'])
    print("Armazém:", pessoa['company']['name'])
    print("---")
```
